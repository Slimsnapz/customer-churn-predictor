**Balancing the dataset**

```python
inputs_unbalanced.shape, targets_unbalanced.shape
```

```python
from imblearn.over_sampling import SMOTE
from imblearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
import joblib



# 1. Spliting raw data, unbalanced data FIRST (to prevent data leakage)
X_train, X_test, y_train, y_test = train_test_split(inputs_unbalanced, targets_unbalanced, test_size=0.2, random_state=42, stratify=targets_unbalanced)
    
```

```python
# Defining the models to be used which are logistic regression and random forest classifier
models = {
    'Logistic_Regression': LogisticRegression(max_iter=1000, random_state=42),
    'Random_Forest': RandomForestClassifier(random_state=42),
}
```

```python
# Dictionary to store our trained pipelines
trained_pipelines = {}

# Training model
for model_name, model_instance in models.items():
    print(f"--- Training {model_name} ---")
    
    # Building the leak-proof pipeline
    pipeline = Pipeline([
        ('scaler', StandardScaler()),
        ('smote', SMOTE(random_state=42)), 
        ('model', model_instance)
    ])
    
    # Training the pipeline
    pipeline.fit(X_train, y_train)

    # Evaluating model
    predictions = pipeline.predict(X_test)
    print(classification_report(y_test, predictions))

    # Saving the pipeline
    filename = f"{model_name.lower()}_pipeline.pkl"
    joblib.dump(pipeline, filename)
    print(f"Saved to {filename}\n")
```

# **Model Selection & Conclusion**

After training both model for classification, The Random forest model significantly does better is handling prediction of customers likely to churn. 

Random Forest achieved a Class 1 Recall of 0.65 and an F1-Score of 0.62, significantly beating the linear Regression Model.

Telecom churn data contains complex, non-linear relationships (e.g., interactions between international plans and total minutes). Tree-based models capture these intersections much better than linear models
